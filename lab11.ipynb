```python
# (1)
import pandas as pd

# 파일경로를 찾고, 변수 file_path에 저장
file_path = './data/score.csv'

# read_csv() 함수로 데이터프레임 변환. 변수 df1에 저장
df1 = pd.read_csv(file_path)
print('칼럼명 : ', df1.columns)
print('인덱스 : ', df1.index)
df1['total'] = df1['kor'] + df1['eng'] + df1['mat']
df1['avg'] = round(df1['total'] / 3,2)
display(df1)
```

```python
# (2)
import pandas as pd

# 파일경로를 찾고, 변수 file_path에 저장
file_path = './data/score_noheader.csv'

# read_csv() 함수로 데이터프레임 변환. 변수 df2에 저장
df2 = pd.read_csv(file_path, header=None)
display(df2)
print('칼럼명 : ', df2.columns)
print('인덱스 : ', df2.index)
print('칼럼명 추가 ---------------')
df2.columns=['name', 'kor', 'eng', 'mat']
print('칼럼명 : ', df2.columns)
display(df2)
print('total과 avg열 추가 ---------------')
df2['total'] = df2['kor'] + df2['eng'] + df2['mat']
df2['avg'] = round(df2['total'] / 3,2)
display(df2)
```

```python
# (3)
import pandas as pd

df3 = pd.read_csv('./data/score_header.csv', header=3)
display(df3)

print('칼럼명 : ', df3.columns)
print('인덱스 : ', df3.index)
df3['total'] = df3['kor'] + df3['eng'] + df3['mat']
df3['avg'] = round(df3['total'] / 3,2)
display(df3)

df1.to_csv("./output/score_result.csv", header=None, index=False)
df2.to_csv("./output/score_result.csv", mode='a', header=None, index=False)
df3.to_csv("./output/score_result.csv", mode='a', header=None, index=False)
```

```python
# (4)
import pandas as pd

df4 = pd.read_excel('./data/score.xlsx')
display(df4)
df4['total'] = df4['kor'] + df4['eng'] + df4['mat']
df4['avg'] = round(df4['total'] / 3,2)
df4 = df4.sort_values(by='total', ascending=False)
display(df4)
df4.to_excel("./output/score_result1.xlsx") 
df4.to_excel("./output/score_result2.xlsx", index=False)
```
