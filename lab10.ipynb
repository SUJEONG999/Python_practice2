```python
import pandas as pd
data = {
    'name':['둘리', '또치', '도우너', '희동이'],
    'kor':[90, 80, 70, 70],
    'eng':[99, 98, 97, 46],
    'mat':[90, 70, 70, 60],
}
df = pd.DataFrame(data)
df['class']=['1반', '2반','1반','2반']
print(df)
```

```python
df.loc[4]=['마이콜',80,80,80,'1반']
print(df)
```

```python
df.set_index(['name'], inplace=True)
df.columns = ['국어', '영어', '수학', '반번호']
print(df)
```

```python
df.loc['마이콜', ['국어','영어','수학']] = 100
df.loc['희동이', '영어'] = 90
print(df)
```

```python
df.reset_index(inplace=True)
print(df)
df.rename(columns={'name':'성명'}, inplace=True)
print(df)
```

```python
df1 = df.sort_values(by= '국어', ascending=False)
print(df1)
df2 = df.sort_values(by= '영어', ascending=True)
print(df2)
```

```python
df['총점'] = df['국어'] + df['영어'] + df['수학']
df.sort_values(by= '총점', ascending=False,inplace=True)
print(df)
```

```python
df3 = df.drop('반번호', axis=1)
print(df3)
print(df)
```

```python
df4 = df.drop(4)
print(df4)
print(df)
```

```python
# 10 : 또 다른 해결 방법
import pandas as pd
import numpy as np
data1 = {'kor':90, 'mat':80}
data2 = {'kor':90, 'eng':70}
data3 = {'kor':90, 'eng':70, 'mat':80}
series1 = pd.Series( data1 )
series2 = pd.Series( data2 )
series3 = pd.Series( data3 )
print(series1+series2+series3)
print("===========================")
series1 = series1.reindex(['kor', 'mat', 'eng'], fill_value=0)
series2 = series2.reindex(['kor', 'eng', 'mat'], fill_value=0)
print(series1+series2+series3)
```
