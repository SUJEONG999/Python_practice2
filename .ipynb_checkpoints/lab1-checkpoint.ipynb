```python
import urllib.request
import urllib.parse
params = urllib.parse.urlencode({'category': '역사', 'page': 25})
url = "http://unico2013.dothome.co.kr/crawling/exercise.php?%s" % params
with urllib.request.urlopen(url) as f:
    print(f.read().decode('utf-8')) 
```

```python
import requests
r = requests.request('get', 'http://unico2013.dothome.co.kr/crawling/exercise.php', params= {'category': '여행', 'page': 100})
r.encoding = 'utf-8'
if r.text :
    print(r.text)
else :
    print('응답된 콘텐츠가 없어요')
```
