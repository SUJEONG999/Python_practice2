```python
import requests
from bs4 import BeautifulSoup
import re
w_title = []
w_summary = []
w_rating = []
for n in range(1,11): # 10페이지라서 
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?page='+str(n))
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    titlelist = soup.select('tr>td>div.challengeInfo > h6 > a' )
    summarylist = soup.select('div.challengeInfo > div.summary')
    ratinglist = soup.select('div.challengeInfo > div.rating_type > strong') 
    for dom in titlelist:
        content=re.sub("[\n\t]", '', dom.text)
        w_title.append(content.strip())
    for dom in summarylist:
        w_summary.append(dom.text)
    for dom in ratinglist:
        w_rating.append(dom.text)
        
print(w_title)
print(w_summary)
print(w_rating)
```
