```python
from selenium import webdriver

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3) 

driver.get("http://gs25.gsretail.com/gscvs/ko/products/event-goods")
import time
time.sleep(2)

readLinks = driver.find_elements_by_css_selector('div > div > ul > li:nth-child(2) > span')
for readlink in readLinks :
    readlink.click()
    time.sleep(1)
```

```python
import requests
from bs4 import BeautifulSoup
import re

goodsname = []
goodsprice = []
while True:
    goodsnamelist = driver.find_elements_by_css_selector('div > div:nth-child(5) > ul > li> div > p.tit')
    goodspricelist = driver.find_elements_by_css_selector('div > div:nth-child(5) > ul > li > div > p.price > span')
            
    for dom in goodsnamelist:
        goodsname.append(dom.text)
    for dom in goodspricelist:
        content=re.sub("[,원]", '', dom.text)
        goodsprice.append(content)

    btn = driver.find_element_by_css_selector('div:nth-child(5) > div > a.next')

    if not btn.get_attribute('onclick') == "goodsPageController.moveControl(1)":
        break
    btn.click()
    time.sleep(1)
```

```python
print(goodsname)
print(goodsprice)
```

```python
commentLength = len(goodsname) 
with open('output/gs25_twotoone.csv', "wt", encoding="utf-8") as f:
    f.write('goodsname,goodsprice\n')  
    for i in range(commentLength):
        f.write(goodsname[i]+","+goodsprice[i]+'\n')
```
