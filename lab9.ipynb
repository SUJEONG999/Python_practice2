```python
import pandas as pd
blood = ['A형', 'B형', 'O형', 'AB형']
st = [34.2, 27.1, 26.7, 11.5]
sr = pd.Series(st, index=blood)
print('[시리즈 sr 출력]')
print(sr)
print('[시리즈 sr에서 인덱싱과 슬라이싱하여 출력(숫자인덱스 사용)]')
a = sr.iloc[0]
print(a)
b = sr.iloc[3]
print(b)
c = sr.iloc[1]
print(c)
d = sr.iloc[1:]
print(d)
```

```python
print('[시리즈 sr에서 인덱싱과 슬라이싱하여 출력(이름인덱스 사용)]')
a = sr.loc['A형']
print(a)
b = sr.loc['AB형']
print(b)
c = sr.loc['B형']
print(c)
d = sr.loc['B형':]
print(d)
```

```python
print('[시리즈 sr의 크기 출력]')
print(len(sr))
print('[시리즈 sr의 인덱스 출력]')
print(sr.index)
print('[시리즈 sr의 데이터 값 출력]')
print(sr.values)
```

```python
data = {
    'name':['듀크1', '듀크2', '듀크3', '듀크4', '듀크5','듀크6', '듀크7'],
    'kor':[90, 80, 70, 70, 60, 70, 90], 
    'eng':[99, 98, 97, 46, 77, 56, 90],
    'mat':[90, 70, 70, 60, 88, 99, 90],
}
df = pd.DataFrame(data)                            
print(df.iloc[0:5])
print(df.tail( ))
print(df.columns)
print(df['name'])
print(df.eng)
print(df[['kor','mat']])
print('loc함수 사용 -----------')
print(df.iloc[3])
print(df.iloc[0,0])
print(df.iloc[3,2])
print(df.iloc[[2,3],2])
print(df.iloc[[2,3],[2,3]])
```

```python
print('iloc함수 사용 -----------')
print(df.loc[3])
print(df.loc[0,'name'])
print(df.loc[3,'eng'])
print(df.loc[[2,3],'eng'])
print(df.loc[[2,3],['eng','mat']])
```
